## Matplolib 

```python
import numpy as np 

from matplotlib import pyplot as plt 

%matplotlib inline

import seaborn as sns
```

##   Adding  Two lines plot the same plot 

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.plot(x,y1)
plt.plot(x,y2)
plt.grid(True)
plt.show()
```

## Subplot

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.plot(x,y1)

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

## Barplot

```python
student={"Bob":87, "Matt":56, "Sam":27}

names=list(student.keys())
values=list(student.values())

plt.bar(names,values)

plt.title("Distribution of marks")
plt.xlabel("Names of student")
plt.ylabel("Marks of student ")
plt.show()
```

## Horizontal Bar plot

```python
plt.barh(names,values,color='g')
plt.grid(True)
plt.show()
```

## scatter plot 

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
plt.scatter(x,a)
plt.show()
```

```python
## changing mark Aesthetics
plt.scatter(x,a, marker='*', c='r', s=100)
```

## Adding two marker in the  same plot 

```python
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,2,0,3,7,3,2]

plt.scatter(x,a, marker='*', c="g",  s=100)
plt.scatter(x,b , marker="*", c="r", s=200)

plt.show()
```

```python
## Adding subplot
x=[10,20,30,40,50,60,70,80,90]
a=[8,1,7,2,0,3,7,3,2]
b=[7,2,5,2,0,3,7,3,2]
plt.subplot(1,2,1)
plt.scatter(x,a, marker='*', c="g",  s=100)

plt.subplot(1,2,2)
plt.scatter(x,b , marker="*", c="r", s=200)
plt.show()
```

## Box plot

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

## Violin plot

```python
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
from numpy import median,mean
plt.violinplot(data, showmedians=True)
plt.show()
```

## pie chart 

```python
fruit=['Apple','orange','Mango','Guava']
quantity=[67,34,100,29]
plt.pie(quantity, labels=fruit)
plt.show()
```

```python
plt.pie(quantity, labels=fruit , autopct= '%0.1f%%',
        colors=['yellow','blue', 'green','red','orange'])
plt.show()
```

## Doughnut - chart

```python
plt.pie(quantity , labels=fruit, radius=2)
plt.pie([1], colors=['w'], radius=1)
plt.show()
```
